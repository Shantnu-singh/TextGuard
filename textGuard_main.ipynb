```python
# Import Libraries
import string
import numpy as np
import pandas as pd
import seaborn as sns
from nltk import PorterStemmer
from nltk import word_tokenize
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords

from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score , confusion_matrix , precision_score

```

```python
name = ['isSpam' , 'sms' , 'a' ,'b' , 'c']
df = pd.read_csv('spam.csv', encoding = 'ISO-8859-1' , names= name)
# df['random'] = df['sms']
df.sample(5)
```

```python
df = df.drop(columns= ['a' , 'b' , 'c'] )
```

```python
df.isnull().sum()
df = df.drop(index=0)
```

```python
df.duplicated().sum()
df = df.drop_duplicates(keep = "first")
df.duplicated().sum()
```

```python
# Lable Encoding of spam function

lb = LabelEncoder()
df['isSpam'] = lb.fit_transform(df['isSpam'])
```

```python

df['isSpam'].value_counts().plot(kind='pie' , autopct = '%.2f')
```

```python
df['sms'].iloc[0]
```

```python
# Text preprocessing
df['sms'] = df['sms'].str.lower() # into lower case
```

```python
# Remove punctuation
exclude = string.punctuation
def remove_pun(text):
    return text.translate(str.maketrans("" , "" , exclude))

df['sms'] = df['sms'].apply(remove_pun)
df['sms'].iloc[:3]
```

```python
# Remove stopWords
stop_words = set(stopwords.words('english'))

def remove_stop_words(text):
    filter_word = [word for word in text.split() if word not in stop_words ]
    return " ".join(filter_word[:])

df['sms'] = df['sms'].apply(remove_stop_words)
df['sms'].iloc[:3]
```

```python
# Stemming of text data 
ps = PorterStemmer()

def stemming(text):
    return " ".join([ps.stem(word) for word in text.split()])

df['sms'] = df['sms'].apply(stemming)
df['sms'].iloc[:3]
```

```python
# Generating Word Cloud 

df['dummy'] = df['sms'].astype(str)
wc_spam = WordCloud(width = 750 , height = 750 , min_font_size = 10 , background_color = 'white')
wc_spam_not = WordCloud(width = 750 , height = 750 , min_font_size = 10 , background_color = 'white')
spam_wc = wc_spam.generate(df[df['isSpam'] == 1]['dummy'].str.cat(sep = ' '))
not_spam_wc = wc_spam_not.generate(df[df['isSpam'] == 0]['dummy'].str.cat(sep = ' '))
```

```python
# Spam-text word-Cloud 
plt.imshow(spam_wc)
```

```python
# Non spam-text wordCloud 
plt.imshow(not_spam_wc)
```

```python
spam_corpus = []
for msg in (df[df['isSpam']==1]['sms'].tolist()):
    words = msg.split()
    for word in words :
        spam_corpus.append(word)
spam_corpus[:3]
```

```python
non_spam_corpus = []
for msg in (df[df['isSpam']==0]['sms'].tolist()):
    words = msg.split()
    for word in words :
        non_spam_corpus.append(word)
```

Model Building

```python
# Seperating Input and labal
# Tf-idf Vectorization

tf_idf = TfidfVectorizer()

X = tf_idf.fit_transform(df['sms']).toarray()
y = df['isSpam']
```

```python
# Train-test split
X_train , X_test , y_train , y_test = train_test_split(X , y, test_size=0.2 , random_state=2)
print(X.shape ,X_train.shape,X_test.shape)
print(y.shape ,y_train.shape,y_test.shape)
```

```python
# ML Classification models

mlb = MultinomialNB()
svc = SVC(kernel= 'sigmoid')
lr = LogisticRegression(solver='liblinear')
rfc = RandomForestClassifier(n_estimators=50 , random_state=2)
knc = KNeighborsClassifier()
abc = AdaBoostClassifier(n_estimators=50 , random_state=2)
bc = BaggingClassifier(n_estimators=50 , random_state=2)
etc = ExtraTreesClassifier(n_estimators=50 , random_state=2)
gbdt = GradientBoostingClassifier(n_estimators=50 , random_state=2)
dtc = DecisionTreeClassifier(max_depth=6)
```

```python
classifier = {
    'LogisticRegression' : lr,
    'MultinomialNB' : mlb,
    'SupportVectorClassifier' : svc,
    'RandomForestClassifier' : rfc,
    'KNeighborsClassifier' : knc,
    'AdaBoostClassifier' : abc,
    "BaggingClassifier" : bc,
    'ExtraTreesClassifier' : etc , 
    'GradientBoostingClassifier' : gbdt, 
    'DecisionTreeClassifier' : dtc
}
```

```python
def apply_model(clf , X_train , y_train ,  X_test , y_test):
    clf.fit(X_train , y_train )
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_true = y_test , y_pred= y_pred)
    preci = precision_score(y_true = y_test , y_pred= y_pred)

    return acc , preci

```

```python
accuracy_sc = []
precision_sc = []

for name,clf in classifier.items():
    acc_current , prec_current = apply_model(clf , X_train , y_train ,  X_test , y_test)

    print("for : " , name)
    print("accuracy score : " , acc_current)
    print("precision score : " , prec_current)
    accuracy_sc.append(acc_current)
    precision_sc.append(prec_current)
```

```python
evaluation_df_tf_idf = pd.DataFrame({"Algorithm" : classifier.keys() , "Accuracy" : accuracy_sc , "Precision" : precision_sc }).sort_values(by = "Precision" , ascending= False)

print(evaluation_df_tf_idf)
```

```python
# Making visualization
sub = plt.subplots(figsize = (15 , 10))

sns.barplot(x= 'Algorithm' , y= 'value' , hue = 'variable',
            data = evaluation_df_tf_idf.melt(id_vars= 'Algorithm') , palette=('salmon' , 'blue'))
plt.tight_layout()
# plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# Bag of Words
bow = CountVectorizer()

X = bow.fit_transform(df['sms']).toarray()
y = df['isSpam']

# Train-test split
X_train , X_test , y_train , y_test = train_test_split(X , y, test_size=0.2 , random_state=2)
print(X.shape ,X_train.shape,X_test.shape)
print(y.shape ,y_train.shape,y_test.shape)

accuracy_sc_bow = []
precision_sc_bow = []

for name,clf in classifier.items():
    acc_current , prec_current = apply_model(clf , X_train , y_train ,  X_test , y_test)

    print("for : " , name)
    print("accuracy score : " , acc_current)
    print("precision score : " , prec_current)
    accuracy_sc_bow.append(acc_current)
    precision_sc_bow.append(prec_current)
```

```python
evaluation_df_bow = pd.DataFrame({"Algorithm" : classifier.keys() , "Accuracy" : accuracy_sc_bow , "Precision" : precision_sc_bow }).sort_values(by = "Precision" , ascending= False)

print(evaluation_df_bow)
```

```python
sub = plt.subplots(figsize = (15 , 10))

sns.barplot(x= 'Algorithm' , y= 'value' , hue = 'variable',
            data = evaluation_df_bow.melt(id_vars= 'Algorithm') , palette=('salmon' , 'blue'))
plt.tight_layout()
# plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# N - Gram
n_gram = CountVectorizer(ngram_range=(1,2) , max_features=10000)

X = n_gram.fit_transform(df['sms']).toarray()
y = df['isSpam']

# Train-test split
X_train , X_test , y_train , y_test = train_test_split(X , y, test_size=0.2 , random_state=2)
print(X.shape ,X_train.shape,X_test.shape)
print(y.shape ,y_train.shape,y_test.shape)

accuracy_sc_n_gram = []
precision_sc_n_gram = []

for name,clf in classifier.items():
    acc_current , prec_current = apply_model(clf , X_train , y_train ,  X_test , y_test)

    print("for : " , name)
    print("accuracy score : " , acc_current)
    print("precision score : " , prec_current)
    accuracy_sc_n_gram.append(acc_current)
    precision_sc_n_gram.append(prec_current)
```
